```python
from datetime import datetime

import pandas as pd
from datetime import datetime as dt
```

```python
df = pd.read_csv('data/data.csv')
df.isnull().sum()
```

```python
def iso_to_epoch_timestamp(iso_ts):
    return dt.fromisoformat(iso_ts.replace('Z', '+00:00')).timestamp()
```

```python
df['end_timestamp'] = df['utc_end_ts'].apply(iso_to_epoch_timestamp)
df = df[["symbol","end_timestamp", "close","high","low","num_trades","total_btc_volume","total_usd_volume",]]
df
```

```python
df["ts"] = pd.to_datetime(df["end_timestamp"], unit='s')
df.set_index("ts", inplace=True)
df
```

```python
full_index = pd.date_range(start=df.index.min(), end=df.index.max(), freq='10s')
df = df.reindex(full_index)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["end_timestamp", "symbol"], inplace=True)
df
```

```python
df["close"].interpolate(inplace=True)
df["high"].interpolate(inplace=True)
df["low"].interpolate(inplace=True)
df["num_trades"].ffill(inplace=True)
df["total_btc_volume"].interpolate(inplace=True)
df["total_usd_volume"].interpolate(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.to_csv("data/features_cleaned.csv", index=False)
```
